# Stacks and Queues (continued)
We know from last class that a Stack is an ordered collection of elements where items are added and removed from the top.

Queues are ordered collections of elements where items are added to the back and removed (and returned) from the front.

```python
#From scrath implementation of stack

class Stack:
    class _Node:
        def __init__(self, datum):
            self.datum = datum
            self.below = None

    def __init__(self):
        self.top = None

    def push(self, value):
        new_node = self._Node(value)
        if not self.top:
            self.top = new_node
        else:
            new_node.below = self.top
            self.top = new_node

    def pop(self):
        if self.top:
            backup = self.top.datum
            self.top = self.top.below
            return backup
        else:
            raise IndexError("Stack is empty")

    # Nice to have methods
    def peek(self):
        # This can be solved in 0(1)
        if self.top:
            return self.top.datum
        else:
            raise IndexError("Stack is empty")

    def size(self):
        # This can be solved in 0(n)
        # Extra challenge: Tweak this Stack class such that this has a worst case time complexity of 0(1)
        current = self.top
        count = 0
        while current:
            count += 1
            current = current.below
        return count

    def is_empty(self):
        # This can be solved in 0(1)
        return self.top is None
```
